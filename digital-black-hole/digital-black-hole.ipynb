```python
DIGIT_LENGTH = 9
DIGIT_RANGE = 10**DIGIT_LENGTH

def drop(x):
    digits=[]
    y=x
    for i in range(DIGIT_LENGTH):
        digits.append(y%10)
        y=y//10
    digits.sort(reverse=True)
    large=0
    for d in digits:
        large=large*10+d
    digits.sort()
    small=0
    for d in digits:
        small=small*10+d
    return large-small

def dropdown(x):
    numbers=[x]
    blackhole=[]
    for i in range(500):
        y=drop(x)
        if y in numbers:
            blackhole=numbers[numbers.index(y):]
            break
        x=y
        numbers.append(x)
    minpos=blackhole.index(min(blackhole))
    blackhole=blackhole[minpos:]+blackhole[0:minpos]
    return numbers,blackhole
```

```python
import matplotlib.pyplot as plt
import random
blackholes=[]
for r in range(0,200):
    x=random.randrange(DIGIT_RANGE)
    z1,z2=dropdown(x)
    z=z1+z2
    for b in blackholes:
        if len(set(b)&set(z2))!=0:
            break
    else:
        blackholes.append(z2)
    plt.plot(range(len(z)),z)
    #plt.plot(range(10-len(z),10),z)
plt.show()
blackholes.sort()
print(blackholes)
```
